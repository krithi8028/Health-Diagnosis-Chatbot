```python
import pandas as pd
import numpy as np
dataset=pd.read_csv('dataset.csv')
description=pd.read_csv('symptom_Description.csv')
precaution=pd.read_csv('symptom_precaution.csv')
severity=pd.read_csv('Symptom-severity.csv')
```

```python
dataset_melted = dataset.melt(id_vars=['Disease'], value_name='Symptom').dropna()
dataset_melted['Symptom'] = dataset_melted['Symptom'].str.strip().str.lower()
```

```python
merged = dataset_melted.merge(severity, left_on='Symptom', right_on='Symptom', how='left')
```

```python
symptom_agg = merged.groupby('Disease')['Symptom'].apply(lambda x: ', '.join(set(x))).reset_index()
```

```python
severity_scores = merged.groupby('Disease')['weight'].sum().reset_index()
severity_scores.rename(columns={'weight': 'Total_Severity'}, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
severity_scores['Normalized_Severity'] = scaler.fit_transform(severity_scores[['Total_Severity']])
```

```python
final_df = symptom_agg.merge(severity_scores, on='Disease', how='left')
final_df = final_df.merge(description, on='Disease', how='left')
```
