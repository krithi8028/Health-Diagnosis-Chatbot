```python
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Combine all symptoms into a single list
all_symptoms = ",".join(df['Symptom']).split(",")

# Count frequency of each symptom
symptom_counts = Counter(all_symptoms)
top_symptoms = pd.DataFrame(symptom_counts.most_common(15), columns=["Symptom", "Count"])

# Plot symptom distribution
plt.figure(figsize=(12, 6))
sns.barplot(y=top_symptoms["Symptom"], x=top_symptoms["Count"], palette="magma")
plt.title("Top 15 Most Common Symptoms")
plt.xlabel("Count")
plt.ylabel("Symptom")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["Normalized_Severity"], bins=20, kde=True, color="red")
plt.title("Distribution of Normalized Severity")
plt.xlabel("Severity Score")
plt.ylabel("Frequency")
plt.show()
```

```python
top_severe_diseases = df.sort_values(by="Normalized_Severity", ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(y=top_severe_diseases["Disease"], x=top_severe_diseases["Normalized_Severity"], palette="cool")
plt.title("Top 10 Most Severe Diseases")
plt.xlabel("Normalized Severity")
plt.ylabel("Disease")
plt.show()
```
